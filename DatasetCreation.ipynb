```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import seaborn as sns
from sklearn.model_selection import train_test_split
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from math import sqrt
import statsmodels.api as sm
import datetime as dt
```

```python
df = pd.read_csv('complete_karimnagar.csv')
```

```python
df_18 = pd.DataFrame([pd.Series([i,j]) for i,j in zip(df['Date'],df['Max Temp (°C)']) if "/18" in i])
df_19 = pd.DataFrame([pd.Series([i,j]) for i,j in zip(df['Date'],df['Max Temp (°C)']) if "2019" in i])
```

```python
df_18[0] = pd.to_datetime(df_18[0],format="%d/%m/%y")
df_18[df_18[0].dt.strftime("%m/%d")=="01/05"].max()
```

```python
df_19[0] = pd.to_datetime(df_19[0])
df_19.head(40)
```

```python
df_20 = pd.DataFrame([pd.Series([i,j]) for i,j in zip(df['Date'],df['Max Temp (°C)']) if "2019" not in i and "2021" not in i and "2022" not in i and "-20" in i or "2020" in i ])
df_20[0] = pd.to_datetime(df_20[0])
df_20.head()
```

```python
df_21 = pd.DataFrame([pd.Series([i,j]) for i,j in zip(df['Date'],df['Max Temp (°C)']) if "-21" in i and "2019" not in i and "2020" not in i and "2022" not in i])
df_21[0] = pd.to_datetime(df_21[0])
df_21.head()
```

```python
df_22 = pd.DataFrame([pd.Series([i,j]) for i,j in zip(df['Date'],df['Max Temp (°C)']) if "-22" in i and "2019" not in i and "2020" not in i and "2021" not in i])
df_22[0] = pd.to_datetime(df_22[0])
df_22.head()
```

```python
df_adil = pd.concat([df_18,df_19,df_20,df_21,df_22],axis=0)
```

```python
df_adil.info()
```

```python
df_adil = df_adil.sort_values(0).reset_index()
df_adil = df_adil.drop(columns=['index', 'level_0'])
df_adil.head()
#drop index first then index and label
```

```python
grp_df = df_adil.groupby(df_adil[0])
len(list(grp_df))
list(grp_df)[0][1].max().to_frame().transpose()
```

```python
df_adil = pd.DataFrame()

for tmp_df in list(grp_df):
  tmp = tmp_df[1].max().to_frame().transpose()
  df_adil = pd.concat([tmp,df_adil],axis=0)
```

```python
df_adil.sort_values(0)[df_adil[0].dt.strftime("%m/%d")=="01/05"].max()
```

```python
df_adil = df_adil[::-1]
```

```python
df_adil = df_adil.rename(columns={0: 'Date', 1: 'Max_temp'})
```
