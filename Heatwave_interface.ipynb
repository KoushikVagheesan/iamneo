```python
import pickle
```

```python
def predict(month,district):
  m = {'January': 0,'February': 1,'March': 2,'April': 3,"May": 4,'June': 5,'July': 6,'August': 7,'September':8,'October': 9,'November': 10,'December': 11}
  pred=0
  if (district == "Adilabad"):
    adil = dict([['January', 34.92781700831105],
 ['February', 35.01989784351883],
 ['March', 34.995538390707466],
 ['April', 35.04709077271487],
 ['May', 35.001473224491676],
 ['June', 35.03943144592105],
 ['July', 34.98890867253362],
 ['August', 35.002150521506856],
 ['September', 34.97248628021466],
 ['October', 34.9699151873101],
 ['November', 34.945459165817105],
 ['December', 34.92603767081828]])

    pred = adil[month]
  elif (district == 'Nizamabad'):
   kham = dict([['January', 35.762470292324736],
 ['February', 35.53308270684371],
 ['March', 35.35399813671122],
 ['April', 35.25236345877777],
 ['May', 35.16283089192487],
 ['June', 35.10065855108332],
 ['July', 35.05655594169566],
 ['August', 35.01010495760978],
 ['September', 34.972054352416286],
 ['October', 34.94229813984692],
 ['November', 34.90775677213778],
 ['December', 34.875183811328]])
   pred = kham[month]
  elif (district == 'Karimnagar'):
    try:
      kari_mod = pickle.load(open('karimnagar_heatwave.pkl','rb'))
      kari_pred = kari_mod.predict('2022-12-01','2023-12-01')[1:]
      pred = kari_pred.values[m[month]]
    except:
      pass
  elif (district == 'Warangal'):
    try:
      kuda_mod = pickle.load(open('warangal_heatwave.pkl','rb'))
      kuda_pred = kuda_mod.predict('2022-12-01','2023-12-01')[1:]
      pred = kuda_pred.values[m[month]]
    except:
      pass
  elif (district == 'Khammam'):
    try:
      niza_mod = pickle.load(open('khammam_heatwave.pkl','rb'))
      niza_pred = niza_mod.predict('2022-12-01','2023-12-01')[1:]
      pred = niza_pred.values[m[month]]
    except:
      pass
  
  return pred
  
```

```python
import gradio as gr

demo = gr.Interface(
    predict,
    [
        gr.Dropdown(["January", "February", "March","April","May","June","July","August","September","October","November","December"]),
        gr.Radio(["Adilabad", "Nizamabad", "Karimnagar","Khammam","Warangal"]),
    ],
    "text",
)

if __name__ == "__main__":
    demo.launch(share=True,debug=True)
```
